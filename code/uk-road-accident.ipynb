#### Data cleaning functions

```python
def get_first_location_data_with_road_type_and_local_authority(local_authority, road_type):
    related_region_accidents = accidents_df[(accidents_df['Local_Authority_(Highway)']==local_authority) & 
             (accidents_df['Road_Type']==road_type) & (~accidents_df['Longitude'].isna())]
    if (len(related_region_accidents)>0):
        return related_region_accidents.head(1)
    else:
        return None

def get_replacement_data_with_road_type_and_local_authority(local_authority, road_type, attribute):
    related_region_accidents = accidents_df[(accidents_df['Local_Authority_(Highway)']==local_authority) & 
             (accidents_df['Road_Type']==road_type) & (~accidents_df[attribute].isna())]
    if (len(related_region_accidents)>0):
        return related_region_accidents.head(1)[attribute]
    else:
        return None

def replace_missing_location_values():
    indexes = accidents_df[accidents_df['Longitude'].isna()].index
    for idx in indexes:
        retrieved = (get_first_location_data_with_road_type_and_local_authority(
            accidents_df.iloc[idx]['Local_Authority_(Highway)'], accidents_df.iloc[idx]['Road_Type']))
        accidents_df.at[idx, 'Longitude'] = retrieved['Longitude']
        accidents_df.at[idx, 'Latitude'] = retrieved['Latitude']
        
def replace_missing_value_based_on_location(attribute):
    indexes = accidents_df[accidents_df[attribute]==-1].index
    for idx in indexes:
        retrieved = (get_replacement_data_with_road_type_and_local_authority(
            accidents_df.iloc[idx]['Local_Authority_(Highway)'], accidents_df.iloc[idx]['Road_Type'], attribute))
        accidents_df.at[idx, attribute] = retrieved
        
def replace_attributes_values_based_on_location():
    columns_to_replace_based_on_location = ["Speed_limit", "Junction_Detail", "Pedestrian_Crossing-Human_Control",
                                        "Pedestrian_Crossing-Physical_Facilities"]
    for attribute in columns_to_replace_based_on_location:
        replace_missing_value_based_on_location(attribute)
        
def replace_attribute_with_most_occuring(dataframe, attributes):
    for attribute in attributes:
        mode = dataframe[attribute].mode()
        dataframe.replace({attribute:{-1:mode}}, inplace=True)
```

```python
def convert_string_time_to_decimal_time(string_time):
    hour_and_minute = re.split(':', string_time)
    hour_and_minute = [int(x) for x in hour_and_minute]
    return hour_and_minute[0] + (hour_and_minute[1]/60.0)

def update_accidents_with_decimal_time():
    for idx in accidents_df['Time'].index:
        decimal_time = convert_string_time_to_decimal_time(accidents_df.at[idx, 'Time'])
        accidents_df.at[idx, 'decimal_time'] = round(decimal_time,3)
        
def update_accidents_with_day_of_year():
    for idx in accidents_df['Date'].index:
        date = accidents_df.at[idx, 'Date']
        date_string = re.split('/',date)
        date_int = [int(x) for x in date_string]
        date_value = datetime.date(date_int[2],date_int[1],date_int[0]).timetuple().tm_yday
        accidents_df.at[idx, 'day_of_year'] = int(date_value)
        
def draw_histogram(data, show_kde, x_axis_title):
    fig, ax = pyplot.subplots(figsize=(8,6))
    # Density Plot with Rug Plot
    sbn.distplot(data, hist = True, kde = show_kde, color = 'darkblue')
    pyplot.xlabel(x_axis_title)
    pyplot.ylabel('Density' if show_kde else 'Count')
    
def create_new_column_with_category_dict(df, category_dict, target_column):
    df[target_column+'_Cat'] = df.apply(lambda x: category_dict[x[target_column]], axis=1)
    
def get_time_of_the_day_category(decimal_time):
    if (decimal_time >=6 and decimal_time <7) or (decimal_time >10 and decimal_time <12):
        return 'Morning'
    elif (decimal_time >=7 and decimal_time <10):
        return 'Morning Rush'
    elif (decimal_time >=12 and decimal_time <16):
        return 'Afternoon'
    elif (decimal_time >=16 and decimal_time <19):
        return 'Evening Rush'
    elif (decimal_time >=19 and decimal_time <20):
        return 'Evening'
    elif (decimal_time >=20 and decimal_time <24):
        return 'Night'
    else:
        return 'Midnight'
    
def draw_count_plot(data, x_column, x_label, y_label, fig_size=(8,6), order=None, hue=None):
    sbn.set(rc = {'figure.figsize':fig_size})
    sbn.countplot(data=data, x=x_column, order=order, color='#1C8BAD', hue=hue )
    pyplot.xlabel(x_label)
    pyplot.ylabel(y_label)
    pyplot.xticks(rotation=45, horizontalalignment='right',fontweight='light',fontsize='large')
    pyplot.show()
    
def set_plot_labels(x_label, y_label, font_size):
    pyplot.xlabel(x_label)
    pyplot.ylabel(y_label)
    pyplot.xticks(rotation=45, horizontalalignment='right', fontweight='light', fontsize=font_size)
    
def get_different_geometry_based_on_col(df, col_name, order=None):
    geometries = []
    unique_items = df[col_name].unique().tolist() if order==None else order
    for v in unique_items:
        sub_df = df[df[col_name]==v]
        sub_geometry = [Point(xy) for xy in zip(sub_df['Longitude'], sub_df['Latitude'])]
        geometries.append(sub_geometry)
    return geometries, unique_items

def draw_map_density_based_on_feature(df, col_name, dimension, order=None):
    geometries, attribute_values = get_different_geometry_based_on_col(df, col_name, order)
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
    cols, rows = dimension[0],dimension[1] 
    fig, axis = pyplot.subplots(nrows=rows, ncols=cols, constrained_layout=True, sharex=True, sharey=True)
    fig.set_size_inches(20, 10)
    count=0
    for irow in range(axis.shape[0]):
        for icol in range(axis.shape[1]):
            axis[irow][icol].axis('off')
            axis[irow][icol].set_xticklabels([])
            axis[irow][icol].set_yticklabels([])
            if count < len(attribute_values):
                gdf = GeoDataFrame(geometry=geometries[count])
                gdf.plot(ax=world[world.name=='United Kingdomn'].plot(ax=axis[irow][icol]), 
                         color='purple', markersize=1, aspect='equal')
                axis[irow][icol].set_title(attribute_values[count], fontsize=17)
                count = count + 1
            else:
                axis[irow][icol].set_visible(False)
    pyplot.subplots_adjust(wspace=0, hspace=0.09)
    pyplot.show()
```

#### Using Apriori for data Mining to get conditions under which accidents are most likely to happen

```python
def prepare_apriori_columns(df, column_list):
    apriori_df = list()
    for col in column_list:
        apriori_df.append(pd.get_dummies(df[col], prefix=col))
    apriori_df = pd.concat(apriori_df, axis=1)
    return apriori_df

def get_conditions_which_road_accidents_most_likely_occur_with_apriori(min_support, min_threshold, 
                                                                       expected_consequent):
    apriori_df = prepare_apriori_columns(acc_vehicle_df,['Speed_limit', 'Weather_Conditions',
                                                       'Accident_Severity', 'Road_Surface_Conditions', 
                           'Sex_of_Driver', 'Vehicle_Manoeuvre', '1st_Road_Class', 'Age_Band_of_Driver', 
                           'Day_of_Week'] )
    rules = apriori(apriori_df, min_support=min_support, use_colnames=True)
    association_rules_df = association_rules(rules, metric='confidence', min_threshold=min_threshold)
    association_rules_df["antecedent_len"] = association_rules_df["antecedents"].apply(lambda x: len(x))
    association_rules_df["consequents_len"] = association_rules_df["consequents"].apply(lambda x: len(x))
    association_rules_df.sort_values(by=['conviction'], ascending=False, inplace=True)
    return association_rules_df[association_rules_df['consequents']=={expected_consequent}][0:5]
```

```python
target_attributes = ['Carriageway_Hazards', 'decimal_time', 'day_of_year', 'Road_Type','Special_Conditions_at_Site',
                    'Pedestrian_Crossing-Human_Control', 'Light_Conditions', 'Day_of_Week', 'Weather_Conditions',
                    'Pedestrian_Crossing-Physical_Facilities', 'Road_Surface_Conditions', 'Longitude', 'Latitude',
                    'Junction_Detail', 'Urban_or_Rural_Area', '1st_Road_Class', 'Speed_limit',
                   'Sex_of_Driver', 'Vehicle_Manoeuvre', 'Age_Band_of_Driver', 'Journey_Purpose_of_Driver',
                   'Was_Vehicle_Left_Hand_Drive?', 'Junction_Location', 'Vehicle_Type', 
                    'Vehicle_Location-Restricted_Lane' ]
```

```python
X, Y = acc_vehicle_casualties_df[target_attributes], acc_vehicle_casualties_df[severity_column]
```

```python
def balance_dataset(X, Y):
    smoteOversampler = SMOTE(sampling_strategy = "not majority")
    randomUndersampler = RandomUnderSampler(sampling_strategy = "auto")
    balance_pipeline =Pipeline([('smote_oversampler', smoteOversampler), 
                                ('random_undersampler', randomUndersampler)])
    return balance_pipeline.fit_resample(X, Y)
```

```python
Y.value_counts()
```

```python
balanced_X, balanced_Y = balance_dataset(X, Y)
```

```python
balanced_Y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(balanced_X, balanced_Y, test_size=0.25, random_state=42)
```

```python
def get_top_features(no_of_features):
    selector = SelectKBest(f_classif, k="all") 
    selector.fit(balanced_X[target_attributes], balanced_Y)
    scores = (selector.scores_)
    argument_sort = list(np.argsort(scores))
    selected_attributes, selected_attribute_scores = [], []
    for i in range(len(target_attributes)-no_of_features,len(target_attributes)):
        value = argument_sort[i]
        selected_attributes.append(target_attributes[value])
        selected_attribute_scores.append(scores[value])
    return selected_attributes, selected_attribute_scores
```

```python
selected_attributes, selected_attributes_score = get_top_features(17)
```
